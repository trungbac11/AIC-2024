```python
from transnetv2 import TransNetV2
import ffmpeg
import cv2
import os
import glob

# Function to save a specific frame
def save_frame(video_cap, frame_number, output_dir):
    # Set to specific frame
    video_cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)

    # Read the frame
    filename = "{:0>6d}.jpg".format(frame_number)
    frame_path = os.path.join(output_dir, filename)
    ret, frame = video_cap.read()
    if ret:
      if not os.path.exists(output_dir):
        os.makedirs(output_dir)

        # Save the frame
      if not cv2.imwrite(frame_path, frame):
            print(f'Failed to save the frame at {frame_path}')
      else:
            print(f'Successfully saved the frame at {frame_path}')
    else:
        print(f'Failed to read the frame at position {frame_number}')

def save_all_frame(output_dir,video_path,scenes):
# Open the video file
  cap = cv2.VideoCapture(video_path)

  # Loop through all scenes and save frames
  for scene in scenes:
      start, end = scene

      # Save the start frame
      save_frame(cap, start, output_dir)

      # Calculate the positions of three evenly spaced middle frames
      frame_step = (end - start) // 2
      middle_frames = start + frame_step

      # Save the three middle frames

      save_frame(cap, middle_frames, output_dir)

      # Save the end frame
      save_frame(cap, end, output_dir)

  # Release the video capture object
  cap.release()
```

```python
video_dir="/PATH/TO/FOLDER/VIDEO" # thay đổi path đến từng folder video lớn
video_files=glob.glob(video_dir+"/*.mp4")
video_files.sort()
model=TransNetV2()
for i,video_path in enumerate(video_files): # nên để chạy hết
    video_frames, single_frame_predictions, all_frame_predictions = model.predict_video(video_path)
    scenes=model.predictions_to_scenes(single_frame_predictions)
    output_dir="/PATH/TO/OUTPUT/FOLDER/" # link tới folder cần chứa
    os.makedirs(output_dir,exist_ok=True) # sẽ tự động tạo thêm folder nếu chưa có
    print(output_dir)
    save_all_frame(output_dir,video_path,scenes)
    print("Done on{}",format(video_path))




```
